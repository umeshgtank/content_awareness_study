```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(15,10)})
```

## Bleu score without Data Poisoning

```python
df = pd.read_csv("data/Bleu/Corrected/msr_paraphrase_analysis_pred_score.txt")
df.head(5)
```

```python
figure, axis = plt.subplots(2, 2)
for i, col in enumerate(list(df.columns.values)):
    sns.scatterplot(x=df.index, y=df[col], ax=axis.flatten()[i])
    
figure.savefig('graphs/bleu.jpeg')
```

```python
figure, axis = plt.subplots(2, 2)
for i, col in enumerate(list(df.columns.values)):
    sns.lineplot(x=df.index, y=df[col], ax=axis.flatten()[i])
```

```python
figure, axis = plt.subplots(2, 2)
for i, col in enumerate(list(df.columns.values)):
    sns.distplot(df[col], ax=axis.flatten()[i])
```
